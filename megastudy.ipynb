```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import re
import json
import pandas as pd
```

```python
def crawling(subject_name):
    # 선생님 수 추출
    teacher_cnt = range(0, len(driver.find_elements_by_css_selector("a.total_g")))
    print(teacher_cnt)
    
    for i in teacher_cnt:
        
        # 해당 과목 선생님 목록 진입
        subject = driver.find_element_by_xpath('//a[contains(text(), "{}")]'.format(subject_name))
        subject.send_keys(Keys.ENTER)
        time.sleep(1)
        
        # 선생님 상세 페이지 진입
        teacher = driver.find_elements_by_css_selector("a.total_g")[i]
        teacher.send_keys(Keys.ENTER)
        time.sleep(1)

        # 개설 강좌 진입
        all_lecture = driver.find_element_by_css_selector("li:nth-child(2) > span > a")
        all_lecture.send_keys(Keys.ENTER)
        time.sleep(1)
        
        btn = driver.find_element_by_css_selector("#iMenuList1")
        btn.send_keys(Keys.ENTER)
        time.sleep(1)
    
        # 강의 제목 크롤링
        titles = driver.find_elements_by_css_selector("p.lstedu_bookinfo--tit > a")

        for title in titles:
            megastudy['title'].append(title.text)
            
            # 제목 수만큼 해당 과목 기입
            megastudy['subject'].append(driver.find_element_by_css_selector("a.lnb_tit--subject").text)
    
        # 선생님 크롤링
        teachers = driver.find_elements_by_css_selector("a.tb_teahome")

        for teacher in teachers:
            megastudy['teacher'].append(teacher.text)

        # 학년 크롤링
        grades = driver.find_elements_by_css_selector("span.lstedu_bookinfo--class__txt")

        for grade in grades:
            megastudy['grade'].append(re.findall('(\[\S*\]).*', grade.text)[0])
            
            
        driver.back()
        time.sleep(1)
        
        driver.back()
        time.sleep(1)
```

```python
driver = webdriver.Chrome('C:/Users/user/Downloads/chromedriver.exe')
```

```python
# 전체 선생님 목록
driver.get("http://www.megastudy.net/teacher_v2/teacher_main.asp")
time.sleep(1)

# 국어 카테고리 진입
subject = '국어' 

megastudy = {'title': [], 'subject': [], 'teacher': [], 'grade': []}

crawling(subject)
```

```python
print(len(megastudy['title']))
print(len(megastudy['teacher']))
print(len(megastudy['subject']))
print(len(megastudy['grade']))
```

```python
df = pd.DataFrame.from_dict(megastudy)
df.to_json('megastudy_kor.json', orient = 'records')
```
